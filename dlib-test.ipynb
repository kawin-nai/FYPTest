```python
import tensorflow as tp
import numpy as np
import os
import sys
import os
import dlib
import glob
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
dlib.DLIB_USE_CUDA = True
```

```python
shape_predictor_path = "./content/pretrained/shape_predictor_5_face_landmarks.dat"
model_path = "./content/pretrained/dlib_face_recognition_resnet_model_v1.dat"

detector = dlib.get_frontal_face_detector()
sp = dlib.shape_predictor(shape_predictor_path)
model = dlib.face_recognition_model_v1(model_path)
```

```python
img_path = "./content/dataset/img_masked.jpg"
img = dlib.load_rgb_image(img_path)
img_detected = detector(img, 1)
print(len(img_detected))
```

```python
img_sp = sp(img, img_detected[0])
img_aligned = dlib.get_face_chip(img, img_sp)
plt.imshow(img_aligned)
```

```python
imgs = []
for (root, dirs, files) in os.walk(".\\content\\celebs\\data\\train"):
    if files:
        for f in files:
            img_path = root + "\\" + f
            img = dlib.load_rgb_image(img_path)
            img_detected = detector(img, 1)
            if img_detected:
                img_sp = sp(img, img_detected[0])
                imgs.append(dlib.get_face_chip(img, img_sp))
                # plt.imshow(img_aligned)
        # print(root)
        # print(files)
```

```python
for idx, i in enumerate(imgs):
  plt.subplot(10, 10, idx+1)
  plt.imshow(i)
plt.show()
```
